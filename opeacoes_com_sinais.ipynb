```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def x(t):

  retorno = []

  for i in range(len(t)):
    if(t[i]<-2):
      retorno.append(0)
    elif(t[i]>4):
      retorno.append(0)
    else:
      retorno.append(t[i]**2)

  return retorno

t = np.linspace(-5,5,500)
y = x(t)

plt.plot(t, y)
```

```python
# x[n-3]
y = x(t-3)

plt.plot(t, y)
```

```python
# x[n+4]
y = x(t+4)

plt.plot(t, y)
```

```python
# x[-n]
y = x(-t)

plt.plot(t, y)
```

```python
# x[-n+3]
y = x(-t+2)

plt.plot(t, y)
```

```python
# x[-n-2]
y = x(-t-2)

plt.plot(t, y)
```

```python
def x(t):

  retorno = []

  for i in range(len(t)):
    if(t[i]<3):
      retorno.append(0)
    else:
      retorno.append(t[i]**2)

  return retorno

t = np.linspace(-5,10,500)
y = x(t)

plt.plot(t, y)
```

```python
# x(1 - t)

y = x(1-t)
plt.plot(t, y)
```

```python
# x(1 - t) + x(2-t)

y = np.add(x(1-t),x(2-t))
plt.plot(t, y)
```

```python
# x(1 - t) + x(2-t)

y = np.multiply(x(1-t),x(2-t))
plt.plot(t, y)
```

```python
# x(3t)

y = x(3*t)
plt.plot(t, y)
```

```python
# x(t/3)

y = x(t/3)
plt.plot(t, y)
```
